```python
moli_smpls=[]
gygi_smpls=[]
mapping = {}
for item in moli_expr.index:
    for jtem in raw_protein_data.columns:
        if item in jtem:
            test = jtem.split("_")[0]
            key = item
            if key in mapping.keys():
                value = mapping[item]
                value.append(jtem)
                mapping[item] = value
            
            else:
                mapping[item] = [jtem]
            moli_smpls.append(item)
            gygi_smpls.append(jtem)
            
```

binarize samples (10% lowest logIC50 is set to sensitive)

```python
ccle_response = pd.read_csv("/home/debian/Mordor/depmap/data/base/dl_data/response/CCLE_response.Paclitaxel.tsv",sep="\t")
ccle_response["response"] = "na"
ccle_response["logIC50"] = np.log(ccle_response["IC50 (uM)"])
ccle_threshold = ccle_response.copy()
ccle_threshold = ccle_threshold.sort_values("logIC50")
ccle_threshold = ccle_threshold.reset_index()
t = int(len(ccle_threshold)/10)
ic_threshold = ccle_threshold.iloc[t]["logIC50"]
len(ccle_response.loc[ccle_response["logIC50"]<ic_threshold]) # number of samples classified as sensitive
```

```python
for item in ccle_response.index:
    if ccle_response.iloc[item]["logIC50"] < ic_threshold:
        ccle_response.at[item, "response"] = "S"
    else:
        ccle_response.at[item, "response"] = "R"
```
